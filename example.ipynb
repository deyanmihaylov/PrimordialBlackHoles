```python
from pbh import PrimordialBlackHole
import numpy as np
import matplotlib.pyplot as plt
```

```python
spacetime = "Kerr"
initial_mass = 100.
final_mass = 1.
initial_momentum = 0.
save_path = True
```

```python
pbh = PrimordialBlackHole(
    spacetime,
    initial_mass, final_mass,
    initial_momentum,
    save_path=save_path,
)
pbh.evolve()
```

```python
steps = np.arange(pbh.path.shape[0])

plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 0])

plt.xlabel(r"step $n$")
plt.ylabel(r"mass M(n)")

plt.show()


plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 1])

plt.xlabel(r"step $n$")
plt.ylabel(r"angular momentum J(n)")

plt.show()


plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 2])

plt.xlabel(r"step $n$")
plt.ylabel(r"a_{*} (n)")

plt.show()


plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 4])

plt.yscale('log')

plt.xlabel(r"step $n$")
plt.ylabel(r"temperature T(n)")

plt.show()
```
